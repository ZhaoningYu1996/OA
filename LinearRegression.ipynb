```python
import numpy as np
```

Normalization

```python
def normalize(data):
	for i in range(0,data.shape[1]-1):
		data[:,i] = ((data[:,i] - np.mean(data[:,i]))/np.std(data[:, i]))
```

Hypothesis

```python
def h(x,theta):
	return np.matmul(x, theta)
```

Cost Function

```python

def cost_function(x, y, theta):
	return ((h(x, theta)-y).T@(h(x, theta)-y))/(2*y.shape[0])
```

Gradient Decent

```python
def gradient_descent(x, y, theta, learning_rate=0.1, num_epochs=10):
	m = x.shape[0]
	J_all = []
	
	for _ in range(num_epochs):
		h_x = h(x, theta)
		cost_ = (1/m)*(x.T@(h_x - y))
		theta = theta - (learning_rate)*cost_
		J_all.append(cost_function(x, y, theta))

	return theta, J_all 
```

```python
def hypothesis(X, beta):
    return [sum(beta[j] * X[i][j] for j in range(len(beta))) for i in range(len(X))]
```

```python
def compute_cost(X, y, beta):
    N = len(y)
    y_pred = hypothesis(X, beta)
    return sum((y_pred[i] - y[i])**2 for i in range(N)) / (2 * N)
```

```python
def gradient_descent(X, y, beta, alpha, iterations):
    N = len(X)
    for _ in range(iterations):
        print(beta)
        y_pred = hypothesis(X, beta)
        for j in range(len(beta)):
            gradient = sum((y_pred[i] - y[i]) * X[i][j] for i in range(N)) / N
            beta[j] -= alpha * gradient
    return beta
```

```python
def multiple_linear_regression(X, y, alpha, iterations):
    beta = [0] * len(X[0])  # Initialize coefficients
    beta = gradient_descent(X, y, beta, alpha, iterations)
    return beta
```

```python
def standardize_features(X):
    means = [sum(column) / len(column) for column in zip(*X)]
    std_devs = [max(1e-7, (sum((x_i - mean) ** 2 for x_i in column) / len(column)) ** 0.5) for column, mean in zip(zip(*X), means)]
    
    # Skip standardization for the first column if it's the intercept term
    return [[x_i if j == 0 else (x_i - means[j]) / std_devs[j] for j, x_i in enumerate(row)] for row in X]
```

```python
X = [[1, 2104, 5, 1, 45], [1, 1416, 3, 2, 40], [1, 1534, 3, 2, 30]]
y = [460, 232, 315]
X = standardize_features(X)
alpha = 0.01
iterations = 100

coefficients = multiple_linear_regression(X, y, alpha, iterations)
print("Coefficients:", coefficients)
```
